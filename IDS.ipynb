```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

from imblearn.over_sampling import BorderlineSMOTE
from xgboost import XGBClassifier

import pandas as pd
import numpy as np
import glob
import os

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report

from imblearn.over_sampling import SMOTE
```

##MERGE DATASETS

```python

# List all CSV files in the directory
all_files = glob.glob(os.path.join(train_path, "*.csv"))

# Initialize an empty list to store individual dataframes
df_list = []

# Read each file and append to the list
for file in all_files:
    # Read the file and add a column to identify the source/attack type
    file_name = os.path.basename(file).replace('.csv', '')
    df = pd.read_csv(file)

    # If the dataset doesn't already have a label column, extract it from filename
    if 'label' not in df.columns:
        # Extract attack type from filename
        if 'benign' in file_name.lower():
            df['label'] = 'Benign'
        elif 'dos' in file_name.lower():
            df['label'] = 'DoS'
        elif 'ddos' in file_name.lower():
            df['label'] = 'DDoS'
        elif 'recon' in file_name.lower():
            df['label'] = 'Recon'
        elif 'mqtt' in file_name.lower():
            df['label'] = 'MQTT'
        elif 'spoofing' in file_name.lower():
            df['label'] = 'Spoofing'
        else:
            df['label'] = 'Unknown'

    df_list.append(df)

# Combine all dataframes into one
combined_train_df = pd.concat(df_list, ignore_index=True)
combined_train_df.head()

# Basic data exploration
print(f"Combined dataset shape: {combined_train_df.shape}")
print(f"Attack distribution:\n{combined_train_df['label'].value_counts()}")
```

```python
# Handle missing values
combined_train_df.fillna(0, inplace=True)

# Convert categorical features if needed
# For example, if you have protocol names as strings
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['Protocol','State'] # Add other categorical columns
# Add any other columns that contain string values
for col in combined_train_df.columns:
    if col != 'label' and combined_train_df[col].dtype == 'object':
        if col not in categorical_cols:
            categorical_cols.append(col)

print(f"Categorical columns detected: {categorical_cols}")

for col in categorical_cols:
    if col in combined_train_df.columns:
        le = LabelEncoder()
        combined_train_df[col] = le.fit_transform(combined_train_df[col])

# Feature scaling
from sklearn.preprocessing import StandardScaler
features = combined_train_df.drop(['label'], axis=1)
labels = combined_train_df['label']

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
X = combined_train_df.drop(['label'], axis=1)
y = combined_train_df['label']

# Scale features
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)

# Encode labels
le = LabelEncoder()
encoded_y = le.fit_transform(y)

# Split into training and validation
X_train, X_val, y_train, y_val = train_test_split(
    scaled_X, encoded_y, test_size=0.2, random_state=42)

# 🛠 3. Handle Imbalance for Neural Network

# Use SMOTE to balance the training set
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

# Calculate class weights (FOR validation/test sets)
classes = np.unique(y_train_res)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train_res)
class_weight_dict = dict(zip(classes, weights))

print("✅ Data ready for training.")
```

#NEURAL NETWORKS


```python


# 🛠 4. Build Neural Network Model

model = Sequential([
    Dense(128, input_dim=X_train_res.shape[1], activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(len(classes), activation='softmax')  # Output layer (multi-class)
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 🛠 5. Train Model

early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    X_train_res, y_train_res,
    validation_data=(X_val, y_val),
    epochs=50,
    batch_size=1024,
    class_weight=class_weight_dict,
    callbacks=[early_stopping],
    verbose=1
)

# 🛠 6. Validate Model

val_preds = model.predict(X_val)
val_preds_labels = np.argmax(val_preds, axis=1)

val_true_labels = le.inverse_transform(y_val)
val_pred_decoded = le.inverse_transform(val_preds_labels)

print("\n📋 Validation Results:")
print(classification_report(val_true_labels, val_pred_decoded))
```

```python
combined_train_df['binary_label'] = combined_train_df['label'].apply(
    lambda x: 'normal' if str(x).lower() == 'benign' else 'intrusion'
)

# Encode as 0 and 1
combined_train_df['label_encoded'] = combined_train_df['binary_label'].map({'normal': 0, 'intrusion': 1})
combined_train_df.drop(columns=['label'], inplace=True)

print(f"Dataset shape: {combined_train_df.shape}")
print(f"Binary label distribution:\n{combined_train_df['binary_label'].value_counts()}")




# Path to your test CSVs
test_path = "/content/drive/MyDrive/IoMT_Dataset/WiFI_and_MQTTattacksCSVtest"
# List all CSV files
test_files = glob.glob(os.path.join(test_path, "*.csv"))

test_df_list = []

for file in test_files:
    file_name = os.path.basename(file).replace('.csv', '')
    df = pd.read_csv(file)

    # Create a 'label' column if not present
    if 'label' not in df.columns:
        if 'benign' in file_name.lower():
            df['label'] = 'Benign'
        elif 'dos' in file_name.lower():
            df['label'] = 'DoS'
        elif 'ddos' in file_name.lower():
            df['label'] = 'DDoS'
        elif 'recon' in file_name.lower():
            df['label'] = 'Recon'
        elif 'mqtt' in file_name.lower():
            df['label'] = 'MQTT'
        elif 'spoofing' in file_name.lower():
            df['label'] = 'Spoofing'
        else:
            df['label'] = 'Unknown'

    test_df_list.append(df)

# Combine all DataFrames
combined_test_df = pd.concat(test_df_list, ignore_index=True)

# Convert to binary labels
combined_test_df['binary_label'] = combined_test_df['label'].apply(
    lambda x: 'normal' if str(x).lower() == 'benign' else 'intrusion'
)

# Encode as 0 and 1
combined_test_df['label_encoded'] = combined_test_df['binary_label'].map({'normal': 0, 'intrusion': 1})

# Optional: Drop original label column if not needed
combined_test_df.drop(columns=['label'], inplace=True)

# Quick check
print(f"Test dataset shape: {combined_test_df.shape}")
print(f"Binary label distribution:\n{combined_test_df['binary_label'].value_counts()}")


```

```python
def clean_features(df):
    # Drop non-numeric columns except 'label_encoded'
    non_numeric = df.select_dtypes(include=['object']).columns.tolist()
    cols_to_drop = [col for col in non_numeric if col not in ['binary_label', 'label']]

    # Drop potential IDs or timestamps (adjust these based on your actual columns)
    id_cols = [col for col in df.columns if 'id' in col.lower() or 'time' in col.lower()]
    cols_to_drop.extend(id_cols)

    # Drop columns with only one unique value
    for col in df.columns:
        if df[col].nunique() <= 1:
            cols_to_drop.append(col)

    # Drop duplicates just in case
    df = df.drop(columns=list(set(cols_to_drop)), errors='ignore')
    df = df.drop_duplicates()

    return df

combined_train_df = clean_features(combined_train_df)
combined_test_df = clean_features(combined_test_df)

# Target column
y_train = combined_train_df['label_encoded']
y_test = combined_test_df['label_encoded']

# Drop label columns from features
X_train = combined_train_df.drop(columns=['label_encoded', 'binary_label', 'label'], errors='ignore')
X_test = combined_test_df.drop(columns=['label_encoded', 'binary_label', 'label'], errors='ignore')

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reshape for RNN input: [samples, timesteps, features]
X_train_rnn = X_train_scaled.reshape((X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_rnn = X_test_scaled.reshape((X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))
```

#Supervised RNN for Binary Classification

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix

# Build RNN Model
rnn_model = Sequential([
    SimpleRNN(64, activation='relu', input_shape=(1, X_train_rnn.shape[2])),
    Dense(1, activation='sigmoid')
])

rnn_model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Train RNN
rnn_model.fit(X_train_rnn, y_train, epochs=10, batch_size=64, validation_data=(X_test_rnn, y_test))

# Evaluate
y_pred_rnn = (rnn_model.predict(X_test_rnn) > 0.5).astype(int)
print("RNN Classification Report:\n", classification_report(y_test, y_pred_rnn))
print("RNN Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rnn))
```

#Recurrent Autoencoder (Unsupervised)

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, RepeatVector, TimeDistributed, Input

# Use only normal samples (label = 0) for training
X_train_auto = X_train_rnn[y_train == 0]

# Define Autoencoder
input_dim = X_train_auto.shape[2]
timesteps = 1

inputs = Input(shape=(timesteps, input_dim))
encoded = LSTM(64, activation='relu', return_sequences=False)(inputs)
repeat = RepeatVector(timesteps)(encoded)
decoded = LSTM(64, activation='relu', return_sequences=True)(repeat)
output = TimeDistributed(Dense(input_dim))(decoded)

autoencoder = Model(inputs, output)
autoencoder.compile(optimizer='adam', loss='mse')

# Train Autoencoder
autoencoder.fit(X_train_auto, X_train_auto, epochs=10, batch_size=64, validation_split=0.1)

# Predict reconstruction on test data
X_test_pred = autoencoder.predict(X_test_rnn)
reconstruction_error = np.mean(np.square(X_test_rnn - X_test_pred), axis=(1, 2))

# Set threshold (e.g., 95th percentile of normal training error)
reconstruction_threshold = np.percentile(
    np.mean(np.square(X_train_auto - autoencoder.predict(X_train_auto)), axis=(1, 2)),
    95
)

# Predict anomalies
y_pred_auto = (reconstruction_error > reconstruction_threshold).astype(int)
print("Autoencoder Classification Report:\n", classification_report(y_test, y_pred_auto))
print("Autoencoder Confusion Matrix:\n", confusion_matrix(y_test, y_pred_auto))
```
